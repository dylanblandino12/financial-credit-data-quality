# Data Quality KPIs

This notebook operationalizes data quality metrics generated in SQL by calculating
impact percentages and severity levels for monitoring purposes.


```python
import pandas as pd

# Base metrics

data = {
    "dimension": ["Completeness", "Validity", "Consistency"],
    "issue_count": [1380, 3, 53538],
    "total_records": [299990, 299990, 299990]
}

df = pd.DataFrame(data)

display(df)
```

```python
#Calculate Impact Percentage

df["impact_pct"] = ((df["issue_count"] / df["total_records"]) * 100).round(3)

display(df)
```

```python
# Severity Classification


def classify_severity(pct):
    if pct >= 10:
        return "High"
    elif pct >= 1:
        return "Medium"
    else:
        return "Low"

df["severity"] = df["impact_pct"].apply(classify_severity)

display(df)
```
